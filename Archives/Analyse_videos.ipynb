# Load Match

```python
championnat =[]
conn = sqlite3.connect('Championnat_U18_GrandEst.sqlite')
cur = conn.cursor()
cur.execute('SELECT * FROM Championnat ')  #WHERE Joueur = ? ', ('MORANDanthony',)
rows = cur.fetchall()
for row in rows:
    line ={}
    line['fichier']= row[1]
    line['Journée'] = row[2]
    line['equipe_dom']=row[3]
    line['equipe_ext']=row[4]
    line['Min']=row[5]
    line['Sec']=row[6]
    line['score_dom']=row[7]
    line['score_ext']=row[8]
    line['Action'] =row[9]
    line['Action_equipe'] =row[10]
    line['Joueur']=row[12]
    championnat.append(line)
conn.close()
```

```python
print("Nombre d'actions du championnat:",len(championnat))
```

```python
df = pd.DataFrame(championnat)
df['Min']=df['Min'].astype(int)
df['Sec']=df['Sec'].astype(int)
df['Temps']=df['Min']+df['Sec']/60
```

# Load Video

```python
lst =glob.glob('D:/Films perso/4_EPINAL U18M-STRASBOURG&SCHILTIGHEIM U18M2/*.MOD')
lst
```

```python
def analyse_video(f):
    video_dict={}
    # video_dict['item']= re.findall('\((\d+)\)',f)
    video_dict['file']=f[lst[0].find('\\')+1:]
    cap = cv.VideoCapture(f)
    # fps = cap.get(cv.CAP_PROP_FPS)      # OpenCV v2.x used "CV_CAP_PROP_FPS"
    # frame_count = int(cap.get(cv.CAP_PROP_FRAME_COUNT))
    # duration = frame_count/fps
    # print('fps = ' + str(fps))
    video_dict['fps']= cap.get(cv.CAP_PROP_FPS)      # OpenCV v2.x used "CV_CAP_PROP_FPS"
    video_dict['frame_count']=  int(cap.get(cv.CAP_PROP_FRAME_COUNT))
    video_dict['duration_sec']=video_dict['frame_count']/video_dict['fps']
    # print('number of frames = ' + str(frame_count))
    # print('duration (S) = ' + str(duration))
    video_dict['minutes']=int(video_dict['duration_sec']/60)
    video_dict['seconds']=int(video_dict['duration_sec']%60)
    # minu = int(duration/60)
    # seco = int(duration%60)
    # print('duration (M:S) = ' + str(minutes) + ':' + str(seconds))
    end_time = datetime.fromtimestamp(os.path.getmtime(f))
    # start_time = end_time -datetime(end_time.year,end_time.month,end_time.day,0,minutes,seconds)
    video_dict['start_time']=end_time -timedelta(minutes=video_dict['minutes'],seconds=video_dict['seconds'])
    video_dict['end_time']=datetime.fromtimestamp(os.path.getmtime(f))
    # print(f'File ending time : {end_time}') 
    # print(f'File starting time : {start_time}')
    cap.release()
    return video_dict
```

```python
videos =  []
for v in lst:
    videos.append(analyse_video(v))
df = pd.DataFrame(videos)
```

```python
df.sort_values(by='start_time', inplace=True)
df.reset_index(inplace=True)
```

```python
df.loc[0,'Delta_Start_End']=df.loc[0,'end_time']-df.loc[0,'start_time']
df['Delta_Break']=timedelta(minutes=0)
for i in range(df.shape[0]-1):
    df.loc[i+1,'Delta_Start_End']=df.loc[i+1,'end_time']-df.loc[i+1,'start_time']
    df.loc[i+1,'Delta_Break']=df.loc[i+1,'start_time']-df.loc[i,'end_time']
```

```python
df['T_start_cumul_forcé']=''
df['Comment']=''
```

```python
def calc_cumul():
    df.loc[0,'T_start_cumul']=0
    for i in range(df.shape[0]-1):
        if df.loc[i+1,'T_start_cumul_forcé']!='':
            df.loc[i+1,'T_start_cumul']=df.loc[i+1,'T_start_cumul_forcé']
        else:
            df.loc[i+1,'T_start_cumul']=df.loc[i,'T_start_cumul']+df.loc[i,'Delta_Start_End'].seconds//60+df.loc[i,'Delta_Start_End'].seconds%60/100*100/60
```

```python
calc_cumul()
```

```python
# Detection mi-temps et forçage du temps 
for i in range(df.shape[0]-1):
    if (df.loc[i+1,'Delta_Break']>timedelta(minutes=8)):
        df.loc[i+1,'Comment']='Mi-temps'
        df.loc[i+1,'T_start_cumul_forcé']=30  
calc_cumul()
```
