```python
import sqlite3
import os
import matplotlib.pyplot as plt
import pandas as pd
import time
import numpy as np
```

```python
# Chargement des matchs dans df
conn = sqlite3.connect('Championnat_U18_GrandEst.sqlite')
cur = conn.cursor()
```

```python
df_classement = pd.read_sql_query("SELECT * FROM Classement", conn)
df_matchs_init = pd.read_sql_query("SELECT * FROM Matchs", conn)
df_matchs = df_matchs_init.drop(df_matchs_init[df_matchs_init['diff_buts']==""].index)
df_déroulé = pd.read_sql_query("SELECT * FROM Championnat", conn)
df_classement[['index','Position','PTS', 'G', 'N', 'P', 'Buts+', 'Buts-', 'Diff', 'J_dom', 'J_ext']].style.hide_index() # N'affiche pas les index des dataframes
```

```python
df_déroulé['Temps']=df_déroulé['Min'].astype(int)+df_déroulé['Sec'].astype(int)/60
df_déroulé['Diff_buts']=df_déroulé['score_dom']-df_déroulé['score_ext']
print(f"Nombre d'actions répertoriées: {df_déroulé.shape[0]}")
```

```python
Liste_equipe = list(df_classement.sort_values(by=['PTS', 'Diff'],ascending=False)['index'])
```

```python
fig,axs = plt.subplots(12, 12,sharex=True, sharey=True,figsize=(21,12), constrained_layout=True,facecolor=(0.5,0.5,0.5))
fig.supylabel('Equipe visiteur (rouge si victoire, vert si défaite)')
fig.supxlabel('Equipe à domicile (vert si victoire, rouge si défaite)')
for i in df_matchs.index:
    journee =int(df_matchs.loc[i]['Journee'])
    equipe1= df_matchs.loc[i]['equipe_dom']
    equipe2= df_matchs.loc[i]['equipe_ext']
    nb1 = Liste_equipe.index(equipe1) # equipe à domicile
    nb2 = Liste_equipe.index(equipe2) # equipe visiteur
    color_threshold = 0.3
    color_factor = df_matchs.loc[i]['diff_buts']/(30/(1-color_threshold)) # Dernier score du match
    #print(df_matchs.loc[i]['fichier'], df_matchs.loc[i]['diff_buts'])
    axs[nb2,nb1].set_xlim([0,60])
    axs[nb2,nb1].set_xticks([0,15,30,45,60])
    axs[nb2,nb1].set_ylim([-25, 25])
    data = df_déroulé[df_déroulé['fichier']==df_matchs.loc[i]['fichier']].sort_values(by=['Temps'])
    axs[nb2, nb1].plot(data['Temps'],data['Diff_buts'],c='orange',linewidth=2)
    axs[nb2, nb1].text(30, 0, df_matchs.loc[i]['diff_buts'],color='white',fontsize='x-large')
    axs[nb2, nb1].grid()
    #
    if color_factor > 0:
        axs[nb2, nb1].set_facecolor((0,color_threshold+color_factor,0))
    elif color_factor < 0:
        axs[nb2, nb1].set_facecolor((color_threshold-color_factor,0,0))
    elif color_factor == 0:
        axs[nb2, nb1].set_facecolor((0.7,0.7,0.7))
    
#
nb = len(Liste_equipe)
# Affichage des noms des équipes
for j in range(nb):
        name = str(Liste_equipe.index(Liste_equipe[j])+1)+"_"+Liste_equipe[j].replace("ENT","").replace("U18M","").replace("ASPTT","")
        axs[nb-1, j].set_xlabel(name[:12],fontsize=8)
        axs[j, 0].set_ylabel(name[:12],fontsize=8)
        axs[j, j].set_facecolor((0.5,0.5,0.5))
# Affichage de la JOURNEE / PREDICTION
for i in range(df_matchs_init.shape[0]):
    Pred="J:"+str(df_matchs_init['Journee'][i])# +"  P:"+str(round(df_matchs['Prediction'][i]))
    axs[Liste_equipe.index(df_matchs_init['equipe_ext'][i]),Liste_equipe.index(df_matchs_init['equipe_dom'][i])].text(10, -15, Pred,color='grey')
titre = "Analyse_Equipes_U18.png"
fig.savefig(titre, dpi=400)
# os.system("Analyse_Equipes_U18.png") #Ouvre l'image
```
