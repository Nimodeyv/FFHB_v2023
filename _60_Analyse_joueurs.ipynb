```python
from collections import defaultdict
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

# Comptabilisation des actions du match (déroulé)

```python
        # 0  ID TEXT NOT NULL PRIMARY KEY UNIQUE, 
        # 1  fichier TEXT,
        # 2  Journée INTEGER,
        # 3  equipe_dom TEXT,
        # 4  equipe_ext TEXT,
        # 5  Min TEXT,
        # 6  Sec TEXT,
        # 7  score_dom INTEGER,
        # 8  score_ext INTEGER,
        # 9  Action TEXT,
        # 10 Action_equipe TEXT,
        # 11 Numero TEXT,
        # 12 Joueur TEXT
```

```python
championnat =[]
conn = sqlite3.connect('Championnat_U18_GrandEst.sqlite')
cur = conn.cursor()
cur.execute('SELECT * FROM Championnat ')  #WHERE Joueur = ? ', ('MORANDanthony',)
rows = cur.fetchall()
for row in rows:
    line ={}
    line['fichier']= row[1]
    line['Journée'] = row[2]
    line['equipe_dom']=row[3]
    line['equipe_ext']=row[4]
    line['Min']=row[5]
    line['Sec']=row[6]
    line['score_dom']=row[7]
    line['score_ext']=row[8]
    line['Action'] =row[9]
    line['Action_equipe'] =row[10]
    line['Joueur']=row[12]
    championnat.append(line)
conn.close()
```

```python
print("Nombre d'actions du championnat:",len(championnat))
```

```python
stat = defaultdict(lambda : defaultdict(int))
for act in championnat:
    #print(act)
    if act['Action']!='TempsMort':
        if act['Action']=='But7m' : 
            stat[act['Joueur']]['But7m']+=1
            stat[act['Joueur']]['Tir']+=1
            stat[act['Joueur']]['But']+=1
        elif act['Action']=='But': 
            stat[act['Joueur']]['But']+=1
            stat[act['Joueur']]['Tir']+=1
        elif act['Action']=='Tir': stat[act['Joueur']]['Tir']+=1
        elif act['Action']=='Avertissement': stat[act['Joueur']]['Avertissement']+=1
        elif act['Action']=='Arrêt': stat[act['Joueur']]['Arrêt']+=1
        elif act['Action']=='2MN': stat[act['Joueur']]['2MN']+=1
        #else: print(act['Action'], 'Action non identifiée ! Check')
        #if 'JV' in act:  stat[joueur]['equipe']=equipe_ext
        #elif 'JR' in act:  stat[joueur]['equipe']=equipe_dom 
        #else: print(act, 'Equipe joueur non identifiée')
#for (j,score) in stat.items():
#    print(j,score)
```

```python
print("Nombre de joueurs du championnat:",len(stat))
```

```python
df = pd.DataFrame(stat)
df = df.transpose()
df.fillna(0, inplace=True)
df['Efficacité%']=round(df['But']/df['Tir']*100)
#df.sort_values(by=['Efficacité%'],ascending=False)[df['But']>50][:20]
```

Classement des Joueurs par nombre de buts

```python
df_buts=df.sort_values(by=['But'],ascending=False)
df_buts['Position'] = df_buts.reset_index().index +1
# df_buts[:30][ 'Position','But', 'Tir', '2MN', 'Avertissement', 'But7m', 'Arrêt', 'Efficacité%']
df_buts[:30].astype(int)
```

#### Classement des Joueurs par efficacité

```python
df_eff=df[df['But']>50].sort_values(by=['Efficacité%'],ascending=False)
df_eff['Position'] = df_eff.reset_index().index +1
df_eff[:30]
```

Classement des Gardiens par nombre d'arrets

```python
df_arret=df.sort_values(by=['Arrêt'],ascending=False)
df_arret['Position'] = df_arret.reset_index().index +1
df_arret[:30].fillna(0).astype(int)
```
